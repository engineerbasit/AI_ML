```python
# importing important libraries 
import tensorflow as tf
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
```

```python
# checking version of libraries
print(tf.__version__)
print(np.__version__)
```

```python
# Loading iris dataset from sklearn datasets
data = datasets.load_iris()
```

```python
# extracting features from data
X = data['data']
# extracting target from data
y = data['target']
```

```python
# Spliting data for training and testing set
xtr,xts, ytr, yts = train_test_split(X,y, test_size = 0.23, random_state = 5)
print(xtr.shape)
print(xts.shape)
print(ytr.shape)
print(yts.shape)
```

## Training a Decision Tree classifier model

```python
DTC_model = DecisionTreeClassifier(max_depth = 3, random_state = 1)
DTC_model.fit(xtr,ytr)
DTC_pred=DTC_model.predict(xts)
DTC_acc = metrics.accuracy_score(DTC_pred,yts)
print("The accuracy of the Decision Tree is : ",DTC_acc)
```

## Training a Linear Discriminant Analysis model

```python
LDA_model = LinearDiscriminantAnalysis(solver='svd')
LDA_model.fit(xtr,ytr)
LDA_pred=LDA_model.predict(xts)
LDA_acc = metrics.accuracy_score(LDA_pred,yts)
print("The accuracy of the Linear Discriminant Analysis is : ",LDA_acc)
```

## Training a Logistic Regression model

```python
LogReg_model = LogisticRegression()
LogReg_model.fit(xtr,ytr)
LogReg_pred=LogReg_model.predict(xts)
LogReg_acc = metrics.accuracy_score(LogReg_pred,yts)
print("The accuracy of the Linear Discriminant Analysis is : ",LogReg_acc)
```
